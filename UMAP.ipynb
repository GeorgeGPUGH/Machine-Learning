```python
import matplotlib.pyplot as plt
import umap
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Create and fit the UMAP model
umap_model = umap.UMAP(n_neighbors=5, min_dist=0.3, metric='correlation')
X_umap = umap_model.fit_transform(X)

# n_neighbors and min_dist are the two key hyperparameters in UMAP and can be adjusted based on the dataset and desired outcome
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y)
plt.title("UMAP on Iris Dataset")
plt.xlabel("UMAP feature 1")
plt.ylabel("UMAP feature 2")
plt.colorbar()
plt.show()
```
